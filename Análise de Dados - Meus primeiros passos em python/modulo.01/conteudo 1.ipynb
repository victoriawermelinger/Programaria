# **1.5 - Qual personagem escolheu você?**

Se você fosse começar um novo curso, como o que está começando agora, como gostaria de ser vista? Qual a imagem que gostaria de passar?

A de pessoa segura, confiante, estudiosa, dedicada?

Infelizmente, nem sempre podemos escolher esse tipo de coisa. Pelo contrário. As personagens escolhem as pessoas. É isso que compartilha a atriz, dubladora e modelo Eli Ferreira em sua palestra:

https://www.youtube.com/watch?v=ApC6C9jeKD0


Já deu para entender que essa aula é um grande convite, né? Queremos convidar você a se despir e deixar toda a sua vulnerabilidade à vista. Aqui você não precisa ser forte o tempo todo, você pode errar (na verdade deve, são com os erros que mais aprendemos a codar!), você pode se sentir pertencente, pois esse espaço foi construído com muito cuidado para ser um local seguro.

Aliás, esse também é um convite para a conexão. Assim como Eli cita bell hooks, por aqui também somos fãs e acreditamos muito no poder do afeto, da rede, da solidariedade e da comunidade como um grande impulso em nossa trajetória. Então se jogue nessa aventura, com toda a sua coragem e vulnerabilidade. Ninguém aqui é perfeite, todes erramos mas, principalmente, todes temos a capacidade de aprender.

Inclusive, é por isso que nosso próximo convite é para assistir a palestra da Carol Dweck:  

https://www.youtube.com/watch?v=_X0mgOOSpLU


É sempre bom relembrar que uma mentalidade de crescimento ajuda a manter a determinação presente no nosso dia a dia, pois sim, você é capaz de aprender algo novo! 💜 Então, caso se depare com um bug, não se esqueça de pensar: "eu ainda não achei o erro". Jamais pense: "eu sou um erro".

Xô, Síndrome de impostore!

Sabemos que não existe um botão que podemos apertar para simplesmente deixarmos a vulnerabilidade fazer parte do dia a dia, assim como incorporar 100% do tempo a mentalidade de crescimento. Por isso, nada como ter palavras de incentivo que façam lembrar que sim, você é capaz de aprender e que você merece fazer parte dessa rede! 

Nada de Síndrome de impostore por aqui! Clique no botão aqui do lado e mande nos tópicos/comentários uma mensagem motivacional para outre astronauta desta missão (e pra você também, né)!

E sempre que estiver sentindo um cheirinho de autossabotagem no ar, volte a esta aula! 

#VamosJuntes e #VamosEmFrente

🌌 Referências para ir ao infinito e além 🌌
