A relação entre aluno e escola é 1:N.

 * 1:N significa que um aluno (1) está relacionado a uma única escola (N). Em outras palavras, um aluno só pode estar matriculado em uma escola de cada vez.

 * A escola (N), por sua vez, pode ter muitos alunos (1) matriculados.

Exemplo:

 * Aluno: João

 * Escola: Colégio Alfa

João (1 aluno) está matriculado no Colégio Alfa (1 escola). O Colégio Alfa (1 escola) tem muitos outros alunos matriculados além de João.

Por que não é N:1?

 * Não faz sentido que um aluno esteja matriculado em várias escolas ao mesmo tempo.
