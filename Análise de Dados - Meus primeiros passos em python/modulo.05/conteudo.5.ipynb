A relao entre aluno e escola  1:N.

 * 1:N significa que um aluno (1) est relacionado a uma nica escola (N). Em outras palavras, um aluno s pode estar matriculado em uma escola de cada vez.

 * A escola (N), por sua vez, pode ter muitos alunos (1) matriculados.

Exemplo:

 * Aluno: Joo

 * Escola: Colgio Alfa

Joo (1 aluno) est matriculado no Colgio Alfa (1 escola). O Colgio Alfa (1 escola) tem muitos outros alunos matriculados alm de Joo.

Por que no  N:1?

 * No faz sentido que um aluno esteja matriculado em vrias escolas ao mesmo tempo.
